```python
import pandas as pd
pd.set_option('display.max_columns', None)
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
mart = pd.read_excel('supermarket_sales.xlsx')
mart.head()
```

```python
mart.describe()
```

```python
mart.columns
```

```python
sns.countplot(x = mart['Product line'])
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x = mart['Product line'])
```

```python
sns.countplot(y = mart['Product line'])
```

```python
sns.countplot(x = mart['Product line'], hue = mart['Gender'])
```

```python
sns.countplot(y = mart['Product line'], hue = mart['Gender'])
```

```python
sns.boxplot(y = 'Total', data = mart, width= 0.2)
```

```python
sns.boxplot(y = 'Total',x = 'Payment', data = mart)
```

```python
sns.boxplot(y = 'Total',x = 'Payment', data = mart, hue = 'Gender')
```

```python
sns.boxplot(y = 'Total',x = 'Payment', data = mart, hue = 'Gender', showmeans = True)
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(data = mart)
```

```python
sns.violinplot(data = mart, y = 'Total')
```

```python
sns.violinplot(data = mart, y = 'Total', x = 'Payment')
```

```python
sns.violinplot(data = mart, y = 'Total', x = 'Payment', hue = 'Gender')
```

```python
sns.violinplot(data = mart, y = 'Total', x = 'Payment', hue = 'Gender', split=True)
```

```python
sns.violinplot(data = mart, y = 'Total', x = 'Payment', hue = 'Gender', split=True, inner = 'quartile')
```

```python
sns.violinplot(data = mart, y = 'Total', x = 'Payment', hue = 'Gender', split=True, inner = 'quartile', bw = 0.1)
```
